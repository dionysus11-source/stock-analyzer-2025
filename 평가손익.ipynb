```python
import pandas as pd
```

```python
# Try different encodings commonly used for Korean text
df = pd.read_csv('TradingHistory.csv', encoding='euc-kr')
```

```python
# astype 없이 문자열 처리
df['평가손익_clean'] = df['평가손익'].str.replace(',', '').str.replace(' ', '').str.strip()

# 숫자로 변환
df['평가손익_clean'] = pd.to_numeric(df['평가손익_clean'], errors='coerce').fillna(0)

# 그룹별 합계 및 정렬
result = df.groupby(['종목명'])['평가손익_clean'].sum().sort_values(ascending=False)

# 세자리마다 쉼표 표시
for stock, value in result.items():
    print(f"{stock}: {value:,.0f}")

```
